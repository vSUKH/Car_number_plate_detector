```python
import numpy as np

def calculate_precision_recall(tp, fp, fn):
    """
    Calculate Precision and Recall given the True Positives, False Positives, and False Negatives.
    """
    precision = [(t / (t + f)) if (t + f) > 0 else 0 for t, f in zip(tp, fp)]
    recall = [(t / (t + n)) if (t + n) > 0 else 0 for t, n in zip(tp, fn)]
    return precision, recall

def calculate_accuracy(tp, fp, fn, tn):
    """
    Calculate accuracy based on True Positives, False Positives, False Negatives, and True Negatives.
    """
    accuracy = [(t + tn_) / (t + f + n + tn_) if (t + f + n + tn_) > 0 else 0
                for t, f, n, tn_ in zip(tp, fp, fn, tn)]
    return accuracy

# Example values (replace with your model's actual outputs)
tp = [148.12, 161, 135.079]  # True Positives
fp = [18.879999999999995, -50, -79.07900000000001]  # False Negatives
fn = [12.879999999999995, 0, 25.920999999999992]  # False Positives
tn = [150,]  # True Negatives (optional)

# Calculate Precision and Recall
precision, recall = calculate_precision_recall(tp, fp, fn)
print("Precision:", precision)
print("Recall:", recall)

# Calculate Accuracy (if applicable, depending on your use case)
accuracy = calculate_accuracy(tp, fp, fn, tn)
print("Accuracy:", accuracy)
```
